```python
# First let's import the packages used in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading in the data
df = pd.read_csv(r'C:\Users\HP\Downloads\archive\movies.csv')
```

```python
# Now let's take a look at the data

df
```

```python
# Need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
df
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
df['Year'] = df['released'].astype(str).str[:4]
df
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#We can see that using this heat map that Gross and Budget have the strongest coorelation
#Gross and Votes have a strong coorelation too
```

```python
# Need to turn numerical values into categorical values in order to include in the correlation heat map

df_numerize = df
for col_name in df_numerize.columns:
        if(df_numerize[col_name].dtype == 'object'):
                df_numerize[col_name] = df_numerize[col_name].astype('category')
                df_numerize[col_name] = df_numerize[col_name].cat.codes
        
df_numerize
```

```python
df
```

```python
#final heat map that includes all features
correlation_matrix = df_numerize.corr(method='pearson')


sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Using this final heat map, we can see that the STRONGEST correlation is the one between BUDGET and GROSS
#Also there exist a strong correlation between GROSS and VOTES, and also between votes and budget
```

```python
# Let's filter and take a look at the ones that have a high correlation (> 0.5)

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values(kind="quicksort")

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
